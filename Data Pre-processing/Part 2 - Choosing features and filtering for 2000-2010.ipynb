```
# Read Lab 1 CSV and store in Pandas Dataframe
df = pd.read_csv("/content/drive/My Drive/Coursework/Spring 2020/CSE 564 - Visualization/Assignments/Lab 2/Datasets/Final_Dataset.csv")
```

## **Process Data**

```
# Filter for 2 years
years = [2000, 2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010]
df_result = df[df['Date'].isin(years)].copy()
df_result = df_result.rename(columns={"Date": "Year"})
df_result
```

```
# Trim Trailing Spaces
columns = ['Country', 'Region']

df[columns] = df[columns].apply(lambda x: x.str.strip())
```

```
# Drop extra categorical columns
df_result.drop(columns=['Population_Density_Categories', 
                      'Health_Expenditure_Govt_Categories', 
                      'Health_Expenditure_Pvt_Categories', 
                      'Health_Expenditure_Indv_Categories', 
                      'Income_Inequality_Categories', 
                      'Employment_Rate_Categories', 
                      'Corruption_Perception_Categories'], inplace=True)
```

```
# Drop Corruption Perception (Since Empty)
df_result.drop(columns=['Corruption_Perception'], inplace=True)
```

```
# Drop Military Expenditure (Since it contains strings)
df_result.drop(columns=['Military_Expenditure'], inplace=True)
```

```
# Delete rows with more than 5 NaN values
df_result.dropna(axis=0, thresh=10, inplace=True)
```

```
# Replace NaN values by 0
df_result.fillna(0, inplace=True)
```

## **Save File**

```
# Save to CSV
df_result.to_csv("/content/drive/My Drive/Coursework/Spring 2020/CSE 564 - Visualization/Project/Final_Dataset.csv", index=False)
```
